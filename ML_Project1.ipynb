2. Performing Linear Regression
Linear Regression aims to find a linear relationship between a Feature (input) and a Target (output). In your notebook:

Feature ($X$$X$): Age
Target ($y$$y$): MonthlyIncome
The algorithm calculates a line of best fit ($y = mx + b$$y = mx + b$) that minimizes the distance between the actual data points and the line. The Coefficient ($m$$m$) represents the predicted increase in income for every year of age, and the Intercept ($b$$b$) is the baseline value.
